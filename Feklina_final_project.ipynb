```python
# calculate profit for each product for each year

query = f"""
SELECT
    sales.product_code,
    sales.fiscal_year,
    SUM(
        (prices.gross_price * (1 - discounts.pre_invoice_discount_pct / 100) - costs.manufacturing_cost) 
        * sales.sold_quantity / 1000
    ) AS total_profit_in_thousands
FROM
(SELECT * FROM fact_sales_monthly WHERE customer_code IS NOT NULL) AS sales
LEFT JOIN fact_gross_price AS prices
    ON prices.product_code = sales.product_code 
    AND prices.fiscal_year = sales.fiscal_year
LEFT JOIN fact_manufacturing_cost AS costs
    ON costs.product_code = sales.product_code 
    AND costs.cost_year = sales.fiscal_year
LEFT JOIN fact_pre_discount AS discounts
    ON discounts.customer_code = sales.customer_code 
    AND discounts.fiscal_year = sales.fiscal_year
GROUP BY
    sales.product_code,
    sales.fiscal_year
ORDER BY
    sales.product_code,
    sales.fiscal_year;
"""
yearly_profit_by_products = pd.read_sql_query(query, con)

#yearly_profit.sample(5)

yearly_profit = pd.pivot_table(
    yearly_profit_by_products,
    values='total_profit_in_thousands',
    index='fiscal_year',
    columns='product_code',
    aggfunc='sum',
    fill_value=None
)

yearly_profit.T
```

### Statistical test
<a id='stat_test'></a>

Now let's find out what statistics say about our data sets. Is there a significant difference between them?

We will try to use Mann-Whitney test and Student's t-test.

First let's try to test our hypothesis using Mann-Whitney test for HDD category.

```python
# Mann-Whitney test for the HDD category

sample_india_HDD = sold_quantity_India_HDD['quant_of_purchases']
sample_na_HDD = sold_quantity_NA_HDD['quant_of_purchases']

mann_whitney = stats.mannwhitneyu(sample_india_HDD, sample_na_HDD)
print(f"Mann-Whitney test: p-value = {mann_whitney.pvalue:.5f}")
```

Since the p-value is greater than 0.05, we cannot say that there is a significant difference between the samples.

Interesting! We assumed that there was a difference.
Let's see what the Student's test shows us.

```python
# look at the distribution

sns.histplot(sample_india_HDD, kde=True, label='India', color='blue', alpha=0.5)
sns.histplot(sample_na_HDD, kde=True, label='NA', color='red', alpha=0.5)
plt.legend()
plt.title("Distribution of sales")
plt.show()
```

```python
# compare dispersions

levene_test = stats.levene(sample_india_HDD, sample_na_HDD)
print(f"Levene test for equal dispersions: p-value = {levene_test.pvalue:.5f}")
```

Since `p-value > 0.05`, the dispersions of distributions are quite close. Now let's run the t-test.

```python
# start the t-test
t_test = stats.ttest_ind(sample_india_HDD, sample_na_HDD)
print(f"T-test results: p-value = {t_test.pvalue:.5f}")

# print results
alpha = 0.05
if t_test.pvalue < alpha:
    print("The difference is statistically significant! Reject H0.")
else:
    print("The difference is not statistically significant! We do not reject H0.")
```
